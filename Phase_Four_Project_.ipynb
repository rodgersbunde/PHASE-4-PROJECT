## Metric of Success

For our time series analysis, we have selected the Mean Absolute Percentage Error (MAPE) as the metric of success to assess the viability of the model. MAPE has been chosen due to its ability to provide weighted error values, where errors are divided by the true values. This characteristic is particularly advantageous in handling outliers effectively. In contrast, Root Mean Square Error (RMSE) solely considers the difference between the real and predicted values, which can be misleading if outliers exist in the data.

By utilizing MAPE, we aim to obtain a comprehensive evaluation of the model's performance that accounts for both the magnitude of errors and the relative proportion they represent. This metric offers a more robust assessment and ensures that outliers do not unduly influence the perception of the model's effectiveness in generating accurate predictions.

Therefore, MAPE has been chosen as the preferred metric to gauge the model's performance in our time series analysis. It provides a reliable indication of the model's ability to forecast accurately, considering the potential presence of outliers in the dataset.

### Coefficient of variation (CV)

CV is a measure of the dispersion of data points around the mean and represents the ratio of the standard deviation to the mean. It allows investors to determine how much volatility, or risk, is assumed in comparison to the amount of return expected from investments.

```python
#calculating and creating ROI column
df['ROI'] = (df['2024-02-29']/ df['2000-01-31'])-1
#df['ROI'] = (df['2025-02-28'] / df['2024-03-31'])-1
```

```python
# Calculate mean and standard deviation along the columns axis (axis=1) for each row
mean_values = df.iloc[:, 9:].mean(axis=1)
std_values = df.iloc[:, 9:].std(axis=1)

# Calculate coefficient of variation (CV) using the formula
cv_values = (std_values / mean_values) * 100

# Add the CV column to the DataFrame
df['CV'] = cv_values
```

##  Time Series Components

A time series is composed of four essential components:
1. `level` - The level component represents the baseline or average value around which the time series fluctuates.
2. `trend` - The trend component depicts the long-term upward or downward movement in the time series data.
3. `seasonality` - Seasonality refers to the recurring patterns or cycles that occur at fixed intervals within the time series.
4. `residuals` - residuals capture the random or unpredictable fluctuations in the data that cannot be explained by the other components.

These components collectively determine the pattern observed in the time series data. Changes in these components lead to variations in the time series pattern.

The presence of seasonality and trend in a time series can significantly impact the forecasted values. These components introduce systematic patterns and directional movements that need to be considered when making predictions. By understanding and accounting for the effects of seasonality and trend, more accurate forecasts can be generated for the time series data.

**Time Series Decomposition**

Breaking the non-stationary time series into its three components—trend, seasonality, and residuals—is indeed a helpful approach for investigating the pattern in the past and aiding in the forecasting of future house values.

```python
# checking for trend, seasonality and residuals in data
# Perform seasonal decomposition
result = seasonal_decompose(z_df['Homevalue'], model='additive', period=12)

# Plot decomposition
plt.figure(figsize=(12, 8))

# Original data
plt.subplot(411)
plt.plot(z_df['Homevalue'], label='Original')
plt.legend(loc='best')

# Trend component
plt.subplot(412)
plt.plot(result.trend, label='Trend')
plt.legend(loc='best')

# Seasonal component
plt.subplot(413)
plt.plot(result.seasonal, label='Seasonal')
plt.legend(loc='best')

# Residual component
plt.subplot(414)
plt.plot(result.resid, label='Residuals')
plt.legend(loc='best')

plt.tight_layout()
plt.show()
```

### Stationarity
Checking the stationarity of our series


```python
# Calculate rolling mean and rolling standard deviation
rolling_mean = z_df['Homevalue'].rolling(window=12).mean()
rolling_std = z_df['Homevalue'].rolling(window=12).std()

# Plot rolling statistics
plt.plot(z_df['Homevalue'], color='blue', label='Original')
plt.plot(rolling_mean, color='red', label='Rolling Mean')
plt.plot(rolling_std, color='black', label='Rolling Std')
plt.title('Rolling Mean & Standard Deviation')
plt.xlabel('Date')
plt.ylabel('Homevalue')
plt.legend()
plt.show()

```

```python
from statsmodels.tsa.stattools import adfuller
# Perform Dickey-Fuller test
result = adfuller(z_df['Homevalue'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:', result[4])
```

- ADF Statistic: This value is -0.0287. In an ADF test, a more negative value indicates stronger evidence for stationarity.


- p-value: This value is 0.9561, which is very high. Typically, a p-value less than 0.05 (or sometimes 0.10) is considered statistically significant. In this case, the high p-value suggests we fail to reject the null hypothesis.


- Critical Values: These are benchmark values used to decide if the ADF statistic is significant. The critical values depend on the chosen significance level (like 1%, 5%, or 10%). For example, the critical value at the 5% significance level is -2.87.

In this case, because the p-value is high, we fail to reject the null hypothesis. There isn't enough statistical evidence to conclude the time series is stationary based on the ADF test.

#### Detrending the series
Detrending a time series is an essential step in analyzing and forecasting data, particularly when dealing with non-stationary series. Differencing is indeed a widely used method for detrending time series data. By taking differences between consecutive observations, differencing helps remove the trend component from the data. It can be thought of as a discrete approximation of the derivative of a function.

```python
# Calculate the first-order difference
z_df['Differenced_Values'] = z_df['Homevalue'].diff()

# Drop the NaN value generated by differencing
z_df.dropna(inplace=True)

plt.plot(z_df.index, z_df['Differenced_Values'])
plt.title('Detrended Time Series')
plt.xlabel('Date')
plt.ylabel('Detrended Values')
plt.show()

# Perform Dickey-Fuller test on the detrended time series data
result_detrended = adfuller(z_df['Differenced_Values'])

# Print the ADF statistic, p-value, and critical values
print('ADF Statistic:', result_detrended[0])
print('p-value:', result_detrended[1])
print('Critical Values:', result_detrended[4])
```

### Seasonal AutoRegressive Integrated Moving Average (SARIMA) Model

Generating Parameter Combinations for SARIMA Model

Preparing parameter combinations for a Seasonal Autoregressive Integrated Moving Average (SARIMA) model. By utilizing the itertools module to create various combinations of parameters for both the non-seasonal (p, d, q) and seasonal (P, D, Q, s) components of the SARIMA model.

```python
import itertools
#Iterate through potential pdq values for total dataset
# Define the p, d and q parameters to take any value between 0 and 2
p = d = q = range(0,2)

# Generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, q and q triplets (use 12 for frequency)

pdqs = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

```

```python
#Finding Best SARIMAX Model Parameters
best = []
for combo in pdq:
    for combos in pdqs:
        try:
            s_model = sm.tsa.statespace.SARIMAX(my_series,
                                            order=combo,
                                            seasonal_order=combos,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            output = s_model.fit()
            best.append([combo, combos, output.aic])

        except:
            continue
```

```python
#"Fitting SARIMA Model using Statsmodels SARIMAX"
sarima_model_2 = sm.tsa.statespace.SARIMAX(my_series,
                                order=(3, 1, 3),
                                seasonal_order=(3, 1, 3, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

sari_results = sarima_model_2.fit()

print(sari_results.summary())
```

## Market Drivers Analysis

Economic indicators such as unemployment rates and mortgage interest rates are pivotal in shaping the real estate market landscape. High unemployment rates may reduce demand for housing due to lower consumer confidence and spending power, potentially leading to lower real estate values. Conversely, low mortgage interest rates can stimulate demand for real estate by making borrowing cheaper, possibly driving up property values. By examining these relationships, the bank can anticipate market shifts and adapt its approaches accordingly.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
unemployment_rate_df = pd.read_csv('UNRATE.csv')
mortgage_rate_df = pd.read_csv('MORTGAGE30US.csv')
housing_values_df = pd.read_csv('Neighborhood_month.csv')

# Convert 'DATE' to datetime and calculate monthly averages for Mortgage Rate
unemployment_rate_df['DATE'] = pd.to_datetime(unemployment_rate_df['DATE'])
mortgage_rate_df['DATE'] = pd.to_datetime(mortgage_rate_df['DATE'])
mortgage_rate_monthly_avg_df = mortgage_rate_df.resample('M', on='DATE').mean().reset_index()

# Preprocess Housing Values dataset
housing_dates_columns = housing_values_df.columns[9:]
housing_long_df = pd.melt(housing_values_df, id_vars=['RegionID'], value_vars=housing_dates_columns, var_name='DATE', value_name='AverageValue')
housing_long_df['DATE'] = pd.to_datetime(housing_long_df['DATE'])
monthly_avg_housing_values_df = housing_long_df.groupby('DATE').mean().reset_index()

# Extract year and month
unemployment_rate_df['Year'] = unemployment_rate_df['DATE'].dt.year
unemployment_rate_df['Month'] = unemployment_rate_df['DATE'].dt.month
mortgage_rate_monthly_avg_df['Year'] = mortgage_rate_monthly_avg_df['DATE'].dt.year
mortgage_rate_monthly_avg_df['Month'] = mortgage_rate_monthly_avg_df['DATE'].dt.month
monthly_avg_housing_values_df['Year'] = monthly_avg_housing_values_df['DATE'].dt.year
monthly_avg_housing_values_df['Month'] = monthly_avg_housing_values_df['DATE'].dt.month

# Merge datasets
merged_df_1 = pd.merge(unemployment_rate_df, mortgage_rate_monthly_avg_df, on=['Year', 'Month'], how='inner', suffixes=('_unemp', '_mort'))
final_merged_df = pd.merge(merged_df_1, monthly_avg_housing_values_df, on=['Year', 'Month'], how='inner')
final_merged_df = final_merged_df.drop(columns=['DATE_unemp', 'DATE_mort']).rename(columns={'DATE': 'Date'})

# Plotting
fig, ax1 = plt.subplots(figsize=(14, 8))
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Unemployment Rate (%)', color=color)
ax1.plot(final_merged_df['Date'], final_merged_df['UNRATE'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Mortgage Rate (%)', color=color)
ax2.plot(final_merged_df['Date'], final_merged_df['MORTGAGE30US'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

ax3 = ax1.twinx()
color = 'tab:green'
ax3.spines['right'].set_position(('outward', 60))
ax3.set_ylabel('Average Housing Value', color=color)
ax3.plot(final_merged_df['Date'], final_merged_df['AverageValue'], color=color)
ax3.tick_params(axis='y', labelcolor=color)

plt.title('Comparison of Unemployment Rate, Mortgage Rate, and Average Housing Value (2000-2024)')
fig.tight_layout()
plt.show()
```
